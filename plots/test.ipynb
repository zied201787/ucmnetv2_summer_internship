```python
import re
import matplotlib.pyplot as plt

# ======= Load log file =======
log_file = 'data2.txt'  # Change this to your log path

with open(log_file, 'r') as f:
    lines = f.readlines()

# ======= Initialize metrics =======
epochs = []
train_losses = []
val_losses = []
mious = []
f1_scores = []
accuracies = []
specificities = []
sensitivities = []

# ======= Parse log line by line =======
for line in lines:
    # Training line
    if 'train: epoch' in line and 'iter:0' in line:
        match = re.search(r'epoch (\d+), iter:0, loss: ([\d\.]+)', line)
        if match:
            epoch = int(match.group(1))
            loss = float(match.group(2))
            epochs.append(epoch)
            train_losses.append(loss)

    # Validation line
    elif 'val epoch:' in line:
        match = re.search(
            r'val epoch: (\d+), loss: ([\d\.]+), miou: ([\d\.]+), f1_or_dsc: ([\d\.]+), accuracy: ([\d\.]+),\s*specificity: ([\d\.]+), sensitivity: ([\d\.]+)',
            line
        )
        if match:
            val_losses.append(float(match.group(2)))
            mious.append(float(match.group(3)))
            f1_scores.append(float(match.group(4)))
            accuracies.append(float(match.group(5)))
            specificities.append(float(match.group(6)))
            sensitivities.append(float(match.group(7)))

# ======= Plotting =======
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(epochs, train_losses, label='Train Loss')
plt.plot(epochs, val_losses, label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(epochs, mious, label='mIoU')
plt.xlabel('Epoch')
plt.ylabel('mIoU')
plt.title('Mean IoU over Epochs')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(epochs, accuracies, label='Accuracy')
plt.plot(epochs, sensitivities, label='Sensitivity')
plt.plot(epochs, specificities, label='Specificity')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.title('Accuracy, Sensitivity, Specificity over Epochs')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(epochs, f1_scores, label='F1 Score (Dice)')
plt.xlabel('Epoch')
plt.ylabel('F1 Score')
plt.title('F1 Score over Epochs')
plt.legend()

plt.tight_layout()
plt.show()
```
